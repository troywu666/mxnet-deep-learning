# 用线性回归函数构建单层神经网络

```python
from mxnet import nd
from mxnet import autograd as ag
```

```python
num_exanples=1000
num_inputs=2
true_w=[2,-3.4]
true_b=4.2
features=nd.random.normal(scale=1,shape=(num_exanples,num_inputs))
labels=features[ : ,0]*true_w[0]+features[ : ,1]*true_w[1]+true_b
noise=nd.random.normal(scale=0.01,shape=(labels.shape))
labels+=noise
```

```python
import random
def data_iter(batch_size,features,labels):
    num_examples=len(features)
    indices=list(range(num_examples))
    random.shuffle(indices)
    for i in range(0,num_examples,batch_size):
        j=nd.array(indices[i:min(i+batch_size,num_examples)])
        yield features.take(j),labels.take(j)##根据索引返回相应函数
```

```python
features.take(nd.array([1]))
```

```python
batch_size=10
for x,y in data_iter(batch_size,features,labels):
    print(x,y)
    break
```

```python
w=nd.random.normal(scale=0.01,shape=(num_inputs,1))
b=nd.zeros(shape=(1,))
```

```python
w.attach_grad()
b.attach_grad()
```

```python
def square_loss(y,y_hat):
    return (y.reshape(y_hat.shape)-y_hat)**2/2

def sgd(params,batch_size,lr):
    for param in params:
        param[:]-=lr*param.grad/batch_size##[:]是将内容传送到新地址
        
def linreg(x,w,b):
    return nd.dot(x,w)+b
```

```python
loss=square_loss
net=linreg
epoch=5
lr=0.03

for i in range(epoch):
    for x,y in data_iter(batch_size,features,labels):
        with ag.record():
            loss(y,net(x,w,b)).backward()
        sgd([w,b],batch_size,lr)        
    l=loss(labels,net(features,w,b))
    print("epoch%d  :  loss=%f"%(i+1,l.mean().asnumpy()))
```

# 用softmax函数构建单层神经网络

```python
import d2lzh as d2l
from mxnet import nd
from mxnet import autograd as ag

batch_size=256
train_iter,test_iter =d2l.load_data_fashion_mnist(batch_size)
```

```python
for x,y in train_iter:
    print(x)
```

```python
num_inputs=28*28
num_outputs=10

w=nd.random.normal(scale=0.01,shape=(num_inputs,num_outputs))
b=nd.zeros(shape=num_outputs)

w.attach_grad()
b.attach_grad()
```

```python
def Softmax(X):
    X_exp=X.exp()
    X_exp_sum=X_exp.sum(axis=1,keepdims=True)
    return X.exp()/X_exp_sum
```

```python
X=nd.random.normal(shape=(2,5))
X_prob=Softmax(X)
X_prob,X_prob.sum(axis=1)
```

```python
def net(X):
    return Softmax(nd.dot(X.reshape(-1,num_inputs),w)+b)
```

```python
help(nd.pick)
```

```python
def cross_entropy(y_hat,y):
    return -nd.pick(y_hat,y).log()
```

```python
def evaluate_accuracy(data_iter,net):
    for x,y in data_iter:
        y_hat=net(x)
    return ((y_hat.argmax(axis=1))==y.astype("float32")).mean().asscalar()
```

```python
evaluate_accuracy(test_iter,net)
```
